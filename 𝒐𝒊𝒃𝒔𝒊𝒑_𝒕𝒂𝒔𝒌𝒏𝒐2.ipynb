```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('Unemployment_Rate_upto_11_2020.csv' )
data.head()
```

```python
data.isnull().sum()
```

```python
data.corr()
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(data.corr(),cmap="coolwarm")
plt.show()
```

# data visualization and estimating unempolyment with different factors

```python
fg=px.bar(data,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',template='plotly',title='State wise Unemployment')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

```python
fg=px.bar(data,x='Region.1',y=' Estimated Unemployment Rate (%)',color='Region',template='plotly',title='Region wise Unemployment')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

analayzing data by box_plot

```python
fg=px.box(data,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',template='plotly',title='State wise Unemployment')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

analyzing with scatter plot

```python
fg=px.scatter(data,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',template='plotly',title='State wise Unemployment')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

```python
fg=px.histogram(data,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',template='plotly',title='State wise Unemployment')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

```python
unemp= data[['Region','Region.1',' Estimated Unemployment Rate (%)']]
fig= px.sunburst(unemp,path=['Region.1','Region'],values=' Estimated Unemployment Rate (%)',width=700, height=700,title='unemployment in India')
fig.show()
```
